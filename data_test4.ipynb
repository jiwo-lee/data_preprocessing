```python
import pandas as pd

df = pd.read_csv("problem.csv", encoding="utf-8")
```

```python
df
```

```python
df['출생시간'] = df['출생시간'].astype('str')
df['출생시간']=df['출생시간'].str.rjust(4, fillchar='0') 
```

```python
df
```

```python
df['출생시간']=pd.to_datetime(df['출생시간'], format='%H%M' , errors='coerce').dt.time
```

```python
df
```

```python
df['생년월일'] = pd.to_datetime(df['년도'], format='%Y%m%d', errors='coerce')
df = df.drop(['년도'], axis=1)
```

```python
df
```

```python
split_df = df['생년월일'].astype('str')
split_df = split_df.str.split('-', expand=True)
split_df.columns = ['년도', '월', '일']
df = pd.concat([df, split_df], axis=1)
```

```python
df
```

```python
df['성별'] = df['성별'].str.replace(r'남.*', '남', regex=True)
# 문자 옆에 .*이면 한번이라도 문자가 있으면, .+면 두번이상 문자가 있으면 바꾼다
```

```python
df['성별']=df['성별'].str.replace(r'여.*', '여', regex=True)
```

```python
df
```

```python
df.to_csv('New_Problem.csv', index=False)
```
