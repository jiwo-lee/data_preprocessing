```python
image = cv2.imread('download2.png') #이미지 불러오기
plt.figure(figsize=(12, 4))
plt.imshow(image)
# 치킨이 파랗게 나온다... 이미지는 화소값이 RGB 순서로 저장되는데 cv2에서는 BGR로 읽어서
```

```python
RGB_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) #이미지를 BGR -> RGB로 변경
plt.figure(figsize=(12, 4))
plt.imshow(RGB_image)
# 치킨이 제대로 된 색상으로 출력된다
```

```python
# R, G, B 채널로 분리
R, G, B = cv2.split(RGB_image)

# 각 채널을 이미지로 표시
plt.figure(figsize=(12, 4)) 

# 빨간 채널
plt.subplot(1, 3, 1) # 1행 3열 중 1번째
plt.imshow(R, cmap='Reds')
plt.title('Red Channel') # 제목
          
# 초록 채널
plt.subplot(1, 3, 2) # 1행 3열 중 2번째
plt.imshow(G, cmap='Greens')
plt.title('Green Channel')
          
# 파란 채널
plt.subplot(1, 3, 3) # 1행 3열 중 3번째
plt.imshow(B, cmap='Blues')
plt.title('Blue Channel') 
```

```python
# 이미지 변환 BGR to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# 이미지 변환 RGB to HSV
image_hsv = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2HSV)

# H, S, V 채널로 분리
H, S, V = cv2.split(image_hsv)

# 각 채널을 그레이스케일 이미지로 표시
plt.figure(figsize=(12, 4))

#색상 채널
plt.subplot(1, 3, 1)
plt.imshow(H, cmap='hsv')
plt.title('Hue Channel')

#채도 채널
plt.subplot(1, 3, 2)
plt.imshow(S, cmap='gray')
plt.title('Saturation Channel')

#명도 채널
plt.subplot(1, 3, 3)
plt.imshow(V, cmap='gray')
plt.title('Value Channel') 

plt.show()
```

```python
image = cv2.imread('zebra.jpg')

# 가우시안 블러 적용 (5x5 크기의 커널) - 주변 픽셀의 평균값
blurred_image = cv2.GaussianBlur(image, (5, 5), 0)
blurred_image = cv2.cvtColor(blurred_image , cv2.COLOR_BGR2RGB)

# 중간값 - 주변 픽셀의 중간값
median_filtered_img = cv2.medianBlur(image, 5)
median_filtered_img = cv2.cvtColor(median_filtered_img , cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 6))

plt.subplot(1,2,1)
plt.imshow(blurred_image )
plt.title('Gaussian Blur')

plt.subplot(1,2,2)
plt.imshow(median_filtered_img )
plt.title('Median Blur')

plt.show()
```

```python
# 이미지 정규화
image = cv2.imread('monariza.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

normalized_gray_image = gray_image/255

plt.figure(figsize=(12, 4))
plt.imshow(normalized_gray_image, cmap='gray')
plt.title('Noramlized')
plt.show()

print(normalized_gray_image)
```

```python
#학습데이터의 양을 늘리기 위해 각도를 바꿈
flipped_image1 = cv2.flip(image, 1)
flipped_image2 = cv2.flip(image, 0)
flipped_image3 = cv2.flip(image, -1)

plt.figure(figsize=(12, 10))

plt.subplot(2,2,1)
plt.imshow(image)
plt.title('Original')

plt.subplot(2,2,2)
plt.imshow(flipped_image1)
plt.title('flip1')

plt.subplot(2,2,3)
plt.imshow(flipped_image2)
plt.title('flip0')

plt.subplot(2,2,4)
plt.imshow(flipped_image3)
plt.title('flip-1')

plt.show()
```

```python
image = cv2.imread('download2.png')
image_resized = cv2.resize(image, (64, 64), interpolation=cv2.INTER_AREA)
#64*64 크기로 리사이즈

x, y, width, height = 10, 10, 44, 44
#시작 좌표(x,y), 가로 길이, 세로 높이

#인덱싱을 이용해서 이미지의 특정부분만 자름
cropped_image = image_resized[y:y+height, x:x+width] 
cropped_image = cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 6))
plt.imshow(cropped_image)
plt.title('crop')

plt.show()
```

```python
# 색반전 = 255 - 이미지
image = cv2.imread('download2.png')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_neg = 255 - image_rgb

plt.figure(figsize=(10, 5))

plt.subplot(1,2,1)
plt.imshow(image_rgb)
plt.title('Original RGB Image')

plt.subplot(1,2,2)
plt.imshow(image_neg)
plt.title('Negated RGB Image')

plt.show()
```

```python
# 조명 변화
lab_image = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2LAB) #2D 배열로 모양

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('Original RGB Image')
          
plt.subplot(1, 2, 2)
plt.imshow(lab_image )
plt.title('Lab Image')
          
plt.show()
```
