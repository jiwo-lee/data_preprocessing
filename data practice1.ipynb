```python
import numpy as np
import pandas as pd
```

```python
s1 = pd.read_csv("데이터/sample_1.csv", encoding = "utf-8")
s2 = pd.read_csv("데이터/sample_2.csv", encoding = "utf-8", low_memory=False)
p1 = pd.read_csv("데이터/patient_1.csv", encoding = "cp949")
p2= pd.read_csv("데이터/patient_2.csv", encoding = "cp949")
p3= pd.read_csv("데이터/patient_3.csv", encoding = "cp949")
```

```python
s1 = s1.dropna()
s2 = s2.dropna()
```

```python
#검사코드, 검사결과값표시, 접수시각, 진료과코드, 성별코드nm
s1.columns
```

```python
s1_dropcol = ['검사코드', '검사결과값(표시)', '진료과코드', '진료과코드NM', '성별코드NM']
s1 = s1.drop(s1_dropcol, axis=1)
s2 = s2.drop(s1_dropcol, axis=1)
```

```python
s1
```

```python
s1['입실일자'] = pd.to_datetime(s1['입실일자'], format = '%Y%m%d', errors='ignore')
s1['퇴실일자'] = pd.to_datetime(s1['퇴실일자'], format = '%Y%m%d', errors='ignore')
s1['접수일자'] = pd.to_datetime(s1['접수일자'], format = '%Y%m%d', errors='ignore')

s2['입실일자'] = pd.to_datetime(s2['입실일자'], format = '%Y%m%d', errors='ignore')
s2['퇴실일자'] = pd.to_datetime(s2['퇴실일자'], format = '%Y%m%d', errors='ignore')
s2['접수일자'] = pd.to_datetime(s2['접수일자'], format = '%Y%m%d', errors='ignore')
```

```python
s1['입원기간'] = s1['퇴실일자'] - s1['입실일자']
s2['입원기간'] = s2['퇴실일자'] - s2['입실일자']
```

```python
s1['연구번호'] = s1['연구번호'].apply(str)
s2['연구번호'] = s2['연구번호'].apply(str)

def ID(data):
    return data[:3] + data[5:]

s1['환자ID'] = s1['연구번호'].apply(ID)
s2['환자ID'] = s2['연구번호'].apply(ID)
```

```python
p1columns = ['연구번호', '입실일자', '입실시각', '검사코드NM', '검사결과값(숫자)', '기준치상한값', '기준치하한값',
       '검사나이', '접수일자', '접수시각', '퇴실일자', '퇴실시각', '환자생년월', '성별코드', '환자ID', '입원기간']
s1 = s1[p1columns]
s2 = s2[p1columns]
```

```python
s1
```

```python
def categrize(data):
    if data >= 80:
        return '80세 이상'
    elif 80 > data >= 70:
        return '70대'
    elif 70 > data >= 60:
        return '60대'
    elif 60 > data >= 50:
        return '50대'
    else:
        return '40대'

s1['검사나이'] = s1['검사나이'].apply(categrize)
s2['검사나이'] = s2['검사나이'].apply(categrize)
```

```python
s1
```

```python
s1 = s1.drop(s1[s1['입실시각'] < 100].index, axis=0)
s1 = s1.drop(s1[s1['퇴실시각'] < 100].index, axis=0)
s1 = s1.drop(s1[s1['접수시각'] < 100].index, axis=0)

s2 = s2.drop(s2[s2['입실시각'] < 100].index, axis=0)
s2 = s2.drop(s2[s2['퇴실시각'] < 100].index, axis=0)
s2 = s2.drop(s2[s2['접수시각'] < 100].index, axis=0)
```

```python
s1['입실시각'] = pd.to_datetime(s1['입실시각'], format='%H%M', errors='raise').dt.time
s1['퇴실시각'] = pd.to_datetime(s1['퇴실시각'], format='%H%M', errors='raise').dt.time
s1['접수시각'] = pd.to_datetime(s1['접수시각'], format='%H%M%S', errors='raise').dt.time

s2['입실시각'] = pd.to_datetime(s2['입실시각'], format='%H%M', errors='raise').dt.time
s2['퇴실시각'] = pd.to_datetime(s2['퇴실시각'], format='%H%M', errors='raise').dt.time
s2['접수시각'] = pd.to_datetime(s2['접수시각'], format='%H%M%S', errors='raise').dt.time
```
