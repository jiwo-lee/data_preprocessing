```python
import pandas as pd
import numpy as np

df = pd.read_csv('program_1.csv', encoding='cp949')
```

```python
df
```

```python
dfNum = df[df.select_dtypes(include = ['number']).columns]
dfObj = df[df.select_dtypes(include = ['object', 'category']).columns]
```

```python
dfNum = dfNum.fillna(dfNum.mean())
```

```python
dfNum
```

```python
dfObj
```

```python
dfNew = pd.concat([dfObj, dfNum], axis = 1)
```

```python
dfNew
```

```python
dfNew['평균'] = (dfNew['프로그램1'] + dfNew['프로그램2'] + dfNew['프로그램3'] + dfNew['프로그램4'])/4
```

```python
dfNew
```

```python
def categorize(data):
    if 0 <= data <= 55:
        return '하'
    elif 55 < data <= 70:
        return '중'
    elif 70 < data:
        return '상'
    else:
        '-'

dfNew['평가분류'] = dfNew['평균'].apply(categorize)
```

```python
dfNew
```

```python
dfGood1 = dfNew[dfNew['평가분류'] == '상']
```

```python
dfGood2 = dfNew[dfNew['평가분류'].str.contains('상', na = 'false')]
```
